```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup

page = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.X9JSGy2w1QK")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
forecast = soup.find(id="seven-day-forecast-body")
forecast
```

```python
days = forecast.find_all(class_="period-name")
days
```

```python
days = [day.text for day in days]
days
```

```python
descs = forecast.find_all(class_="short-desc")
descs
```

```python
descs = [desc.text for desc in descs]
descs 
```

```python
temps = forecast.find_all(class_="temp")
temps
```

```python
temps = [temp.text for temp in temps]
temps
```

```python
temp_celsius = []
for i in temps:
    temp_celsius.append(round(int(i.split()[1]) - 32 / 1.8, 1))
print(temp_celsius)
```

```python
df = pd.DataFrame({
    'Day': days,
    'Description': descs,
    'Temp in ° C': temp_celsius
})
df 
```
